# 関数

```python
month=[str(x).zfill(2) for x in np.arange(1,13,1)]
```

```python
seibun=['TN','TP','TK']
seibunlabel=['10a当たりのTN','10a当たりのTP','10a当たりのTK']
```

```python
def stacked_barplot_fertilizer(targetdata, savedirectory, colormap=None):
  if colormap!=None:
    cm = plt.get_cmap(colormap)
    cls=[cm(x/len(seibun)) for x in np.arange(0,len(seibun),1)]
    colors=dict(zip(seibun,cls))
  else:
    colors=None

  for i in range(len(targetdata)):
    ncol=1
    nrow=1
    figx=ncol*8
    figy=nrow*4

    fig, ax=plt.subplots(nrow, ncol, figsize=(figx,figy),tight_layout=True)
    plt.rcParams["font.size"] = 15

    title=targetdata.at[i,'品種']+'：'+targetdata.at[i,'肥料タイプ']

    bottom_data = pd.Series(np.zeros(len(month)), index=month)
    data=targetdata.loc[i,:]
    for k in range(len(seibun)):
      color=colors[seibun[k]]
      col=[seibun[k]+'_'+x for x in month]
      mdata=data.loc[col].astype(int)
      rects=ax.bar(month, mdata, bottom=bottom_data,color=color)
      bottom_data += mdata.values

    ax.set_ylim([llim, hlim])
    #legend
    ax.set_xlabel('月')
    ax.set_ylabel('肥料成分投下量')
    ax.set_title(title)
    ax.minorticks_on()
    ax.grid(which='major', ls='-',alpha=0.5)
    ax.grid(which='minor', ls='--', alpha=0.2)

    plt.legend(seibun, bbox_to_anchor=(1.25, 1), loc='upper right', fontsize=15)

    #text
    rects.datavalues = [np.nan if x == 0 else x for x in bottom_data]
    ax.bar_label(rects,label_type='edge')

    #save
    targetfarmer_directory=savedirectory
    if not os.path.exists(targetfarmer_directory):
      os.mkdir(targetfarmer_directory)
    savename=targetfarmer_directory+title+'.png'
    plt.savefig(savename)
    plt.cla()
    plt.clf()
    plt.close('all')
```
